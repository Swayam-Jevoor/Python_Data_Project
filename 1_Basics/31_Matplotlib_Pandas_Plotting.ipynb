```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Bar Chart - Counts of Job Postings

```python
job_counts = df['job_title_short'].value_counts()

job_counts
```

```python
plt.bar(job_counts.index, job_counts.values)
```

```python
job_counts.plot(kind = 'bar')
```

```python
job_counts.plot(kind = 'line')
```

```python
plt.bar(job_counts.index, job_counts.values)
plt.ylabel('Count of Job Postings')
plt.xlabel('')
plt.title('Job Postings by Job Title')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
df[['job_posted_date', 'salary_year_avg']].dropna(subset = ['salary_year_avg'])
```

```python
df.plot(x = 'job_posted_date', y = 'salary_year_avg', kind = 'line')
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month

df['job_posted_month']
```

```python
df
```

```python
job_posted_month = df.groupby('job_posted_month')['salary_year_avg'].mean()

plt.plot(job_posted_month.index, job_posted_month.values)
plt.title('Salary per Job Postings')
plt.show()
```

```python
monthly_salary = df.groupby('job_posted_month')['salary_year_avg'].mean()

plt.plot(monthly_salary.index, monthly_salary)
plt.title('Salary per Job Postings')
plt.xlabel('Month')
plt.ylabel('Average Salary')
plt.show()
```
