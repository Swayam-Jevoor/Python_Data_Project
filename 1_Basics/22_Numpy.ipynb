```python
import random

random.randint(50000, 100000)
```

```python
salary_list = [random.randint(50000, 100000) for _ in range(1_000_000)]
```

```python
salary_list
```

```python
import statistics

statistics.mean(salary_list)
```

```python
%%timeit

statistics.mean(salary_list)
```

```python
import numpy as np
```

```python
%%timeit

np.mean(salary_list)
```

```python
# Floats and Ints can be added together

# Job titles
job_titles = np.array(['Data Analyst', 'Data Scientist', 'Data Engineer', 'Machine Learning Engineer', 'AI Engineer'])

# Base salaries
base_salaries = np.array([60000, 80000, 75000, 90000, np.nan])

# Bonus rates
bonus_rates = np.array([0.05, 0.1, 0.08, 0.12, np.nan])
```

```python
total_salaries = base_salaries * (1 + bonus_rates)

total_salaries
```

```python
np.mean(total_salaries)
```
