```python
count = 1

while count < 5:
  print(count)
  count += 1
```

```python
years_experience = {
    'Adam':       3,
    'Bob':        6,
    'Charlotte':  4,
    'Diana':      3,
    'Eve':        7
}
```

```python
# Find an applicant that has more than 5 years of experience. Return only first applicant

years = 0
index = 0

years_list = list(years_experience.items())

while years < 5:
  key, years = years_list[index]
  if years >= 5:
    print(key, 'has', years, 'years of experience')
  index += 1
```
