```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Bar Chart - Counts of Job Postings

```python
job_counts = df['job_title_short'].value_counts()
```

```python
job_counts
```

```python
# Adding a title to the bar chart

import matplotlib.pyplot as plt

plt.bar(job_counts.index, job_counts)
plt.title('Postings by Job Title')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation = 45, ha = 'right')  # Rotate x-axis labels by 45 degrees; ha is horizontal alignement
plt.show()
```
