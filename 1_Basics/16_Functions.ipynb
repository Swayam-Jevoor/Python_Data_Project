```python
base_salary = 100000
bonus_rate = 0.1

# Calculate total_salary of base_salary * (1 + bonus_rate)

def calculate_salary():
  total_salary = base_salary * (1 + bonus_rate)

  return total_salary
```

```python
calculate_salary()
```

```python
# Calculate total_salary of base_salary * (1 + bonus_rate)

def calculate_salary(base_salary, bonus_rate):
  total_salary = base_salary * (1 + bonus_rate)

  return total_salary
```

```python
calculate_salary(base_salary, bonus_rate)
```

```python
calculate_salary(100000, 0.1)
```

```python
# Calculate total_salary of base_salary * (1 + bonus_rate)

def calculate_salary(base_salary, bonus_rate = 0.1):
  total_salary = base_salary * (1 + bonus_rate)

  return total_salary
```

```python
salary_1 = 100000
rate_1 = 0.1

calculate_salary(salary_1)
```

```python
salary_1 = 100000
rate_1 = 0.1

calculate_salary(200000)
```

```python
salary_1 = 100000
rate_1 = 0.1

calculate_salary(salary_1, rate_1)
```
