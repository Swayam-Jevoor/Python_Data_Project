## Identity Operators

```python
salary_zoltar = 100_000
salary_strixxi = 100_000

salary_zoltar == salary_strixxi
```

```python
salary_zoltar = 100_000
salary_strixxi = 100_000

salary_zoltar is salary_strixxi

# is operator checks whether two variables are using the same memory location for the value that they are holding
```

```python
id(salary_zoltar)
```

```python
id(salary_strixxi)
```

```python
id(salary_zoltar) == id(salary_strixxi)
```

```python
core_skills = ['Python', 'R', 'SQL']
zoltars_skills = core_skills
```

```python
zoltars_skills is core_skills
```
